### ISTOGRAMMI

```python
fig, ax = plt.subplots(figsize = (6,3))

#ISTOGRAMMA DELLE FREQUENZE ASSOLUTE
ax.hist(fuel_consumption, bins = 50)
ax.set_xlabel("Consumi l/100Km")
ax.set_ylabel("NÂ° di auto")
plt.show()
```

```python
image = plt.imread("./Zucca_particolare.PNG")
image.shape
```

```python
red_component = image[:, :, 0]
green_component = image[:, :, 1]
blue_component = image[:, :, 2]
transparent_component = image[:, :, :3]
```

```python
fig, ax = plt.subplots(figsize = (6,6))
ax.imshow(red_component, cmap="Reds_r")
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (6,3))

ax.hist(red_component.flatten(), bins = 50)
ax.set_xlabel("intensitÃ  del rosso")
ax.set_ylabel("Frequenza assoluta")
plt.show()
```

## SOTTOGRAFICI

```python
fig, axs = plt.subplots(nrows=4, ncols=2, figsize = (16,16))

axs[0,0].imshow(red_component, cmap = "Reds_r")
axs[0,1].hist(red_component.flatten(), bins = 50)
axs[0,0].set_title("Componente rossa")
axs[0,1].set_title("IntensitÃ  colore")

axs[1,0].imshow(green_component, cmap = "Greens_r")
axs[1,1].hist(green_component.flatten(), bins = 50)
axs[1,0].set_title("Componente verde")
axs[1,1].set_title("IntensitÃ  colore")

axs[2,0].imshow(blue_component, cmap = "Blues_r")
axs[2,1].hist(blue_component.flatten(), bins = 50)
axs[2,0].set_title("Componente blu")
axs[2,1].set_title("IntensitÃ  colore")

axs[3,0].imshow(transparent_component, cmap = "Greys_r")
axs[3,1].hist(transparent_component.flatten(), bins = 50)
axs[3,0].set_title("Componente trasparente")
axs[3,1].set_title("IntensitÃ  colore")
plt.show()
```
