Calcolare in quante persone sono nate lo stesso giorno



```python
import numpy as np
```

```python
classi = 1000
alunni = 25
scuola = np.random.randint(1,366, (classi,alunni))
scuola
```

```python
cont = 0
for a in range(classi):
    if len(np.unique(scuola[a])) < alunni:
        cont+=1

print(cont)
    
```
