```python
import numpy as np
import matplotlib.pyplot as plt
```

0 = rabbia

1 = disgusto

2 = paura

3 = felicità

4 = tristezza

5 = sorpresa

6 = neutro

```python
f = open("./fer2013.csv", "r")
lines = f.readlines()
f.close
```

```python
emotions = []
images = []

for line in lines[1:]:
    emotion, pixels, _ = line.split(",")
    emotions.append(int(emotion))
    image = np.array([int(val) for val in pixels.split(" ")])
    images.append(image)
```

```python
emotions = np.array(emotions)
images = np.array(images)
```

```python
fig, ax = plt.subplots(figsize = (3,3))
ax.imshow(images[1299].reshape((48,48)), cmap = "Greys_r")
plt.show
```

```python
emotions_dict = {0:"Rabbia",1:"Disgusto",2:"Paura",3:"Felicità",4:"Tristezza",5:"Sorprea",6:"Neutro",}
```

```python
fig, axs = plt.subplots(nrows=4, ncols=2, figsize=(6,9))
for emotion, ax in enumerate(axs.reshape(-1)):
    if emotion <= 6:
        array_bool = emotions == emotion
        images_emotion = images[array_bool] # contiene le immagini che hanno una determinata emotions
        i = np.random.randint(0, len(images_emotion))
        ax.imshow(images_emotion[i].reshape((48,48)), cmap = "Greys_r")
        ax.set_title(emotions_dict[emotion])
    ax.axis("off")
plt.show()
```
