```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_json("./neowise.json")
data
```

```python
data[["orbit_class","period_yr"]].groupby("orbit_class").mean()
```

```python
fig, ax = plt.subplots(figsize = (10,5))

colors = ["k","g","b","y","r","c","m"]

filter = data[data["period_yr"] < 40]
x = data[data["period_yr"] < 40]["moid_au"].values
y = data[data["period_yr"] < 40]["period_yr"].values
ax.scatter(x, y, s = 3, c=colors[1])
            
ax.set_xlabel("moid_au")
ax.set_ylabel("period_yr")

ax.grid()
plt.show()
```
