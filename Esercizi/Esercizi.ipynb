### Es_007
Disegna un bersaglio 10×10: le cornici concentriche hanno valori crescenti verso il centro (quella più esterna ha valori pari a 1).

```python
n = 10
array = np.ones((n,n), dtype=np.int8)
```

```python
for a in range(1,int(n/2)):
    array[a:-a,a:-a] += 1

plt.imshow(array, cmap= "Reds")
```

```python
n_lanci = 1000
n_partite = 10
a = np.random.randint(1,7,(n_partite, n_lanci))
b = np.random.randint(1,7,(n_partite, n_lanci))
```

```python
punti_a = 0
punti_b = 0

punti_a = np.count_nonzero(a > b) + (np.count_nonzero(a == b)/2)
punti_b = np.count_nonzero(b > a) + (np.count_nonzero(a == b)/2)


print(f"Punti A = {punti_a}\nPunti B = {punti_b}")
```

```python
if punti_a > punti_b + n_lanci/2:
    print("A sta barando")
elif punti_b > punti_a + n_lanci/2:
    print("B sta barando")
else: print("Nessuno sta barando")
```
